```python
import random

n = 500
suma = 0
for i in range(n):
    d1 = random.randint(1,6)
    d2 = random.randint(1,6)

    prod =  d2*d2
    #print(f'{d1} x {d2} = {prod}')
    suma += prod
    
valor_esperado_aprox = suma/n
print(f'Valor aproximado de la suma es de {valor_esperado_aprox}')
```
