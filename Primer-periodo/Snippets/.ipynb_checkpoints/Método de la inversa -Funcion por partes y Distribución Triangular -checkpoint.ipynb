```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def funcion_partes(n):
    r = np.random.rand(n)
    x = np.zeros(n)
    
    for i in range(0,n):
        if 0< r[i] < 0.005:
            x[i] = np.sqrt(2*r[i]) - 1/10
        else:
            x[i] = -10*np.log(201/200 - r[i])
    return x

n = 200000
x = funcion_partes(n)

c = 500
plt.hist(x, bins = c)
plt.show()

print('valor esperado aprox: ', x.mean())
    
```

 ## Funcion triangular (naive)

```python
def funcion_triangular(a,b,c,n):
    r = np.random.rand(n)
    x = np.zeros(n)
    
    for i in range(0,n):
        if 0< r[i] < (c-a)/(b-a):
            x[i] = a + np.sqrt(r[i]*(b-a)*(c-a))
        else:
            x[i] = b - np.sqrt((1-r[i])*(b-a)*(b-c))
    return x

n = 200000
x = funcion_triangular(0,100,80,n)
c = 500
plt.hist(x, bins = c)
plt.show()
```
