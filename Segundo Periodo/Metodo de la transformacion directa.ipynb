# Generacion de numeros aleatorios de la distribucion normal

# Metodo de Box-Muller

```python
import numpy as np
import matplotlib.pyplot as plt
def normal_box_mueller(n, miu, sigma):
    r1 = np.random.rand(n)
    r2 = np.random.rand(n)
    x = miu + sigma *np.sqrt(-2*np.log(1-r1)) * np.cos(2*np.pi*r2)
    return x

x = normal_box_mueller(1000000,100,150)
print(x)

plt.hist(x, bins = 50)
plt.show()
```
